```python
CreditGrade_order=['AA','A','B','C','D','E','HR']
CreditGrade_order1= pd.api.types.CategoricalDtype(ordered= True , categories = CreditGrade_order)
df['CreditGrade']=df['CreditGrade'].astype(CreditGrade_order1)
```

```python
base_color=sns.color_palette()[0]
sns.violinplot(data=df,x= 'CreditGrade', y= 'BorrowerRate', color=base_color );
plt.xlabel('Credit Grade')
plt.ylabel('Borrowing rate')
plt.title("Credit Grade v/s Borrowing Rate");
#Comparing Credit Grade with Borrowing Rate to the relation 
#We can see that the better the rating lesser is the borrowing rate
```

#### We can see that the better the rating lesser is the borrowing rate

### Relation between Qualitative and Quantitative variables

```python
base_color=sns.color_palette()[0]
sns.boxplot(data=df_1,x= 'IncomeRange', y= 'LoanOriginalAmount', color=base_color );
plt.xticks(rotation=20);
plt.xlabel('Income Range')
plt.ylabel('Loan Amount')
plt.title("Income range v/s Loan Amount");
#Comparing Income range with Loan amount
#We observe more the income more is the loan amount
```

#### We observe more the income more is the loan amount

```python
df_2 = df[(df['StatedMonthlyIncome'] <= 10000) ]
df_2 = df_2[df_2['IncomeRange'] != 'Not displayed']


plt.figure(figsize = [15, 8])
base_color = sns.color_palette()[0]

# left plot: violin plot
plt.subplot(1, 2, 1)
ax1 = sns.violinplot(data = df_2, x = 'IncomeRange', y = 'LoanOriginalAmount', color = base_color)
plt.xticks(rotation = 15);
plt.title('Income Range vs Loan Amount')

# right plot: box plot
plt.subplot(1, 2, 2)
sns.boxplot(data = df_2, x = 'IncomeRange', y = 'LoanOriginalAmount', color = base_color)
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot
plt.xticks(rotation = 15);
plt.title('Income Range vs Loan Amount')
```

### Relation between Catogorical Variables

```python

plt.figure(figsize = [12, 8])
ct_counts = df.groupby(['IsBorrowerHomeowner', 'LoanStatus']).size()
ct_counts = ct_counts.reset_index(name = 'count')
ct_counts = ct_counts.pivot(index = 'LoanStatus', columns = 'IsBorrowerHomeowner', values = 'count')
sns.heatmap(ct_counts, annot = True, fmt = 'd');
plt.title('Count of Applicants w.r.t LoanStatus and HomeownerStatus')
#We see that most of the borrowers lie in Completed and Current stage 
#with almost same proportion of Home owner and Non Home owners
```
